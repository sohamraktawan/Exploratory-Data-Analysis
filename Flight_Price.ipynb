# Exploratory Data Analysis of Flight Price Data from Kaggle

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
business = pd.read_csv("./datasets/flight_price_prediction/business.csv")
economy = pd.read_csv("./datasets/flight_price_prediction/economy.csv")
```

```python
business.head()
```

```python
economy.head()
```

### Adding new column "Class" and concatenated to form df

```python
business["Class"] = "business"
economy["Class"] = "economy"

df = pd.concat([business, economy], axis = 0)
```

```python
df.head()
```

### Cleaning "stop" column

```python
df["stop"].value_counts()
```

```python
df["stop"] = df["stop"].str.strip("\n\t\t\t\t\t\t\t\t\t\t\t\t\n\t\t\t\t\t\t\t\t\t\t\t\t")
df["stop"] = df["stop"].str.replace("\n\t\t\t\t\t\t\t\t\t\t\t\t", " ")
df["stop"].value_counts()
```

```python
# Create Number of Stops column
df["Number of Stops"] = df["stop"].str.extract(r'(\d+)').fillna(0).astype(int)

# Create Stop Location column
df["Stop Location"] = df["stop"].str.extract(r'Via (\w+)')[0]
df["Stop Location"] = df["Stop Location"].where(df["stop"].str.contains("Via"), np.nan)
```

```python
df["Stop Location"].value_counts()
```

```python
df.drop("stop", axis = 1, inplace=True)
```

```python
df.head()
```

### Converting price to int

```python
df["price"] = df["price"].str.replace(",", "").astype("int64")
df["price"].head()
```

```python
df.info()
```

### Extracting hour from dept_time

```python
df["dept_hour"] = df["dep_time"].str.split(":").str[0].astype(int)
df["arr_hour"] = df["arr_time"].str.split(":").str[0].astype(int)
```

```python
df.head()
```

```python
df.drop(["dep_time", "arr_time"], axis=1, inplace=True)
```

```python
df.head()
```

### Removing unnecesary columns

```python
df.drop(["ch_code", "num_code"], axis=1, inplace=True)
```

```python
df.head()
```

### Creating new features Day ,  Month , Year based on date column

```python
# Split the date column into Day, Month, Year
df[['Day', 'Month', 'Year']] = df['date'].str.split('-', expand=True)

# Convert Day, Month, Year columns to integers (if needed)
df['Day'] = df['Day'].astype(int)
df['Month'] = df['Month'].astype(int)
df['Year'] = df['Year'].astype(int)
```

```python
df.head()
```

```python
df.drop("date", axis=1, inplace=True)
```

```python
df.head()
```

```python
df["time_taken"].value_counts()
```

### Convert time_taken to time taken in hours

```python
# Function to convert time_taken to total hours
def convert_to_hours(time_str):
    # Extract hours and minutes using str.extract with regex
    hours = pd.to_numeric(time_str.str.extract(r'(\d+)h')[0], errors='coerce')
    minutes = pd.to_numeric(time_str.str.extract(r'(\d+)m')[0], errors='coerce')
    # Convert minutes to hours and sum
    total_hours = hours + (minutes / 60)
    return total_hours

# Apply the conversion to create time_taken_hours column
df["time_taken_hours"] = convert_to_hours(df["time_taken"])
```

```python
df.head()
```

```python
df = df.dropna(subset=["time_taken_hours"])
```

```python
df.drop("time_taken", axis=1, inplace=True)
```

### Multivariate Analysis

### Airline v/s price divided by Class

```python
# Create the plot first
ax = df.groupby(["airline", "Class"])["price"].mean().unstack().plot(kind="bar")

# Set the size of the figure
ax.figure.set_size_inches(10, 6)  # Width = 10 inches, Height = 6 inches

# Annotate each bar with its value
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.2f'),
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 9), textcoords='offset points')
```

```python
plt.figure(figsize=(10,8))
plt.title("Average prices for Economy class")
sns.heatmap(df[df["Class"] == "economy"].groupby(["from","to"])["price"].mean().unstack(), annot=True, fmt='.2f', cmap="Reds")
```

```python
plt.figure(figsize=(10,8))
plt.title("Average prices for Business class")
sns.heatmap(df[df["Class"] == "business"].groupby(["from","to"])["price"].mean().unstack(), annot=True, fmt='.2f', cmap="Reds")
```

```python
plt.figure(figsize=(20,10))
plt.title("Average prices for Departure & Arrival time (Economy Class)")
sns.heatmap(df[df["Class"] == "economy"].groupby(["dept_hour","arr_hour"])["price"].mean().unstack(), cmap="Blues")
```

```python
plt.figure(figsize=(20,10))
plt.title("Average prices for Departure & Arrival time (Economy Class)")
sns.heatmap(df[df["Class"] == "business"].groupby(["dept_hour","arr_hour"])["price"].mean().unstack(), cmap="Blues")
```
