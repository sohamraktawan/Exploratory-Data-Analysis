```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.style.use('ggplot')
```

```python
train = pd.read_csv("./datasets/black_friday/train.csv")
```

```python
test = pd.read_csv("./datasets/black_friday/test.csv")
```

### Ordinal Encoding on Age Column

```python
df["Age"] = df["Age"].map({"0-17":1, "18-25":2, "26-35":3, "36-45":4, "46-50":5,"51-55":6, "55+":7})
```

```python
df.head()
```

```python
df["Gender"] = df["Gender"].map({"M" : 0, "F" : 1})
```

```python
df.info()
```

```python
df["Stay_In_Current_City_Years"] = df["Stay_In_Current_City_Years"].map({"1": 1, "2":2, "3":3, "4+":4})
df["Stay_In_Current_City_Years"].head()
```

### One hot encoding for City Category Column

```python
df = pd.get_dummies(df, columns=['City_Category'], drop_first=True)
```

```python
df.info()
```

```python
df["User_ID"] = df["User_ID"].astype("str")
sns.heatmap(df.select_dtypes(include=["int64", "float64"]).corr())
```

Insights:

1) City Category B has significantly more purchases
2) People who stay in current city for 1 year -> significantly more purchases
3) Unmarried people and Males have significantly more purchases and more purchase value
4) 26-35 is the age group with significantly higher purchases
5) Occupation classes 0, 4 and 7 have significantly higher purchases
6) Mean purchase value is more or less the same across all classes
7) Product Category 5 is the most common primary category, top 5 - 5, 1, 8, 11, 2
8) Bottom Product Categories - 9, 17, 19, 14, 20
9) Product Category 5 - 8 - 16 is the most common in purchases -> with the highest number of purchases
10) Top 5 Products - P00265242 , P00025442, P00110742, P00112142, P00046742
11) Bottom 5 Products - P00359842, P0099542, P00294942, P00074842, P00062442\
12) Products with high price and very low sales - P00074542, P00075042, P00038842, P00013542, P00060842 - (Type 1)
13) Products with low price and high sales - P00371644, P00372445, P00370853, P00375436, P00370293 - (Type 2)
14) Type 1 has more Males and Type 2 has more Females
15) Occupation 9 has more sales from Females than Males
16) Occupation 10 has high number of people in age group 0-17
17) Occupation 13 has high number of old people
18) Occupation 4 is comprised of people mostly from 18-25 age group, which is the most purchases occupation
