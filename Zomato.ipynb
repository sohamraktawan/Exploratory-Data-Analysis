### Areas with lower number of votes and higher mean rating

```python
grouped_df = df[df["Country"] == "India"].groupby('Locality').agg(
    total_votes=('Votes', 'sum'),
    avg_rating=('Aggregate rating', 'mean')
).reset_index()
```

```python
grouped_df = grouped_df[(grouped_df["total_votes"] < grouped_df["total_votes"].mean()) & (grouped_df["avg_rating"] > grouped_df["avg_rating"].mean())]

sorted_df = grouped_df.sort_values(by='total_votes', ascending=False)  # Change to `ascending=False` for descending order

sorted_df
```
