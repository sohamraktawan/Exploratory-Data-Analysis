```python
df = pd.read_csv("./datasets/zomato/zomato.csv", encoding="latin-1")
```

### Merging to make "Country" column

```python
df = pd.merge(left=df, right=ct, on="Country Code", how="left")
```

```python
df.head()
```

```python
df.count()
```

### Checking Null Values

```python
df.isnull().sum()
```

### Drop Null values as proportion insignificant in dataset

```python
df["Cuisines"].dropna(inplace=True)
```

```python
df.isnull().sum()
```

### Top 5 Countries by Number of Restaurants

```python
plt.figure(figsize=(10, 10))
plt.pie(df["Country"].value_counts().nlargest(5), labels=df["Country"].value_counts().nlargest(5).index, autopct='%.0f%%')
```

### Creating new Feature of "Number of Cuisines" based on "Cuisines"

```python
df.dropna(inplace=True)
df['Number of Cuisines'] = df['Cuisines'].apply(lambda x: len(x.split(', ')))
```

```python
df.head()
```

```python
plt.figure(figsize=(10, 10))
plt.title("Count of each Number of Cuisines distributed over Price Range")
sns.countplot(x='Number of Cuisines', hue='Price range', data=df)
```

### Distribution of each Price Range

### Count of each Price Range distributed over Rating
